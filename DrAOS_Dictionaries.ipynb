```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year" : 1964
}
print(thisdict)

print("length of the dictionary is:" , len(thisdict))
#using iterative variable
print('Print Dict with iterative variable')
for x, y in thisdict.items():
      print(x, y)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)

thisdict.pop("model")
print(thisdict)

thisdict.popitem()
print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = dict(thisdict)
print(mydict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```

```python
thisdict = dict(brand="Ford", model="Mustang", year=1964)
print(thisdict)
```

```python
C= dict()
name= ['cse', 'ece', 'cse', 'ece']
for i in name:
  C[i]= C.get(i,0) +1
print(C)
```

```python
C= dict()
print("enter the text")
line=input()
words= line.split()
print(words)
print('counting...')

for i in words:
  C[i]= C.get(i,0) +1
print(C)

```
